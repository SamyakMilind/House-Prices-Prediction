```python
house_data.drop_duplicates(inplace = True)    # Drop duplicates [if any]
house_data.shape
```

```python
# Checking the statistical summary
house_data.describe()
```

```python
#format the date
d =[]
for i in house_data['date'].values:
    d.append(i[:4])
    
house_data['date'] = d

# convert everything to same datatype
for i in house_data.columns:
    house_data[i]=house_data[i].astype(float)
    
#make a new column age of the house  
house_data['age'] = house_data['date'] - house_data['yr_built']

#calculate the total years of renovation
house_data['renov_age'] = np.abs(house_data['yr_renovated'] - house_data['yr_built'])
house_data['renov_age'] = house_data.renov_age.apply(lambda x: x if len(str(int(x)))==2 else 0.0)

#remove unwanted columns like yr_built, date, id
house_data.drop(['id','date', 'yr_built', 'yr_renovated'], axis=1, inplace=True)
house_data.head()
```

```python
# Univariate Analysis 
columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 
          'sqft_above', 'sqft_basement', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'age', 'renov_age']

for each in columns:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13,5))
    # BOX PLOT
    # IQR = q3-q1
    # Any value greater than q3 + 1.5(IQR) or lesser than q1 - 1.5(IQR) is considered an Outlier  
    sns.boxplot(x=each, data=house_data, ax=ax1)
    ax1.set_xlabel(each, fontsize=10)
    ax1.set_ylabel(each, fontsize=10)
    ax1.tick_params(labelsize=10)

    
    # HISTOGRAM
    # Outliers present if the graph is rightly skewed
    ax2.hist(house_data[each])
    ax2.set_xlabel(each, fontsize=10)
    ax2.tick_params(labelsize=10)
```

```python
plt.figure(figsize = (20,10))          # Finding Correlation using Heat-Map
sns.heatmap(house_data.corr(),
           annot = True,
           linewidth = .5,
           cmap = "PiYG",
           fmt='.3f')
```

```python
def RemoveOutliers(df, col_name):
    q1 = df[col_name].quantile(.25)
    q3 = df[col_name].quantile(.75)
    iqr = q3 - q1
    upperlimit = q3 + (1.5 * iqr)
    lowerlimit = q1 - (1.5 * iqr)
    a = df[col_name] > upperlimit
    b = df[col_name] < lowerlimit
    df[col_name] = np.where(a, upperlimit, df[col_name])
    df[col_name] = np.where(b, lowerlimit, df[col_name])
    return df
```

```python
for i in house_data.columns:
    RemoveOutliers(house_data, i)
```

```python
from scipy.stats import zscore
scaled_house_data = house_data.apply(zscore)
scaled_house_data.head()
```

```python
from sklearn.model_selection import train_test_split

XTrain, XTest, yTrain, yTest = train_test_split(house_data.drop('price', axis=1), house_data['price'], test_size = .3,
                                               random_state = 8) # random state is used to maintain consistency
XTrain.shape, XTest.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() # Create a linear regression model
model.fit(XTrain, yTrain)
```

```python
y_pred = model.predict(XTest)
```

```python
from sklearn.metrics import mean_squared_error
# Evaluate the model
mse = mean_squared_error(yTest, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
print("train Accuracy",model.score(XTrain,yTrain))
print("test Accuracy",model.score(XTest,yTest))
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=yTest, y=y_pred)
plt.plot([min(yTest), max(yTest)], [min(y_pred), max(y_pred)], color='red', linewidth=2)  # Plotting the regression line
plt.title('Actual vs. Predicted Values with Regression Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```
